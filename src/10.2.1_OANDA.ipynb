```python
import AdaBoostVectorBacktester as AB
```

```python
abbt = AB.AdaBoostVectorBacktester('USD_JPY', '2002-5-6',
                                    '2009-12-31', 10000)
```

```python
abbt.run_strategy('2002-5-6', '2008-12-31',
                   '2009-1-1', '2009-12-31')
```

```python
abbt.plot_results()
```

```python
abbt.optimize_parameters()
```

```python
abbt.run_strategy('2002-5-6', '2008-12-31',
                   '2009-1-1', '2009-12-31',
                  lags=3,
                  momentum=21,
                  sma=10,
                  volatility=10,
                  min_window=14,
                  max_window=14)
```

```python
abbt.plot_results()
```
