```python
import AdaBoostVectorBacktester as AB
```

```python
abbt = AB.AdaBoostVectorBacktester('EUR_USD',
                                   '2020-06-08', '2020-06-13', 1)
```

```python
abbt.run_strategy(
    '2020-06-08', '2020-06-11 11:00:00',  
    '2020-06-11 11:10:00', '2020-06-13')
```

```python
abbt.plot_results()
```

```python
abbt.optimize_parameters()
```

```python

abbt.run_strategy('2020-06-08', '2020-06-11 11:00:00',
                  '2020-06-11 11:10:00', '2020-06-13',
                  lags=3,
                  momentum=5,
                  sma=10,
                  volatility=20,
                  min_window=14,
                  max_window=14)
```

```python
abbt.plot_results()
```
