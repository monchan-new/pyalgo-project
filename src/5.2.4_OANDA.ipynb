```python
import ScikitVectorBacktester as SCI
```

```python
scibt = SCI.ScikitVectorBacktester('USD_JPY', '2002-5-6',
                                    '2010-12-31', 10000, model='logistic')
```

```python
scibt.run_strategy('2002-5-6', '2008-12-31',
                   '2009-1-1', '2009-12-31')
```

```python
scibt.plot_results()
```

```python
scibt.optimize_parameters((3,51,1),
                          '2002-5-6','2008-12-31',
                          '2002-5-6','2008-12-31')
```

```python
scibt.run_strategy('2002-5-6', '2008-12-31',
                   '2009-1-1', '2009-12-31', lags=22)
```

```python
scibt.plot_results()
```
