```python
import FixDate_ScikitVectorBacktester as SCI
```

```python
scibt = SCI.ScikitVectorBacktester('USD_JPY', '2002-5-6',
                                    '2010-12-31', 10000, model='logistic')
```

```python
scibt.optimize_parameters((3,51,1),
                          '2002-5-6','2005-12-31',
                          '2002-5-6','2005-12-31')
```

```python
scibt.run_strategy('2002-5-6', '2005-12-31',
                   '2006-1-1', '2009-12-31', lags=3)
```

```python

scibt.plot_results()
```
