```python
import NeuralNetworkVectorBacktester as NN
```

```python
nnbt = NN.NeuralNetworkVectorBacktester('USD_JPY', '2002-5-6',
                                    '2009-12-31', 10000)
```

```python
nnbt.run_strategy('2002-5-6', '2005-12-31',
                   '2006-1-1', '2009-12-31')
```

```python
nnbt.plot_results()
```

```python
nnbt.optimize_parameters(start_in='2002-5-6',end_in='2005-12-31',
                          start_out='2002-5-6',end_out='2005-12-31')
```

```python
nnbt.run_strategy('2002-5-6', '2005-12-31',
                   '2006-1-1', '2009-12-31', 
                   lags=7, momentum=10, sma=50, ema=10)
```

```python
nnbt.plot_results()
```
