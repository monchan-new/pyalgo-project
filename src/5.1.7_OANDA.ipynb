```python
import LRVectorBacktester as LR
```

```python
lrbt = LR.LRVectorBacktester('USD_JPY', '2000-1-1', '2010-1-1', 10000, 0)
```

```python
lrbt.run_strategy('2000-1-1', '2005-1-1',
                  '2005-1-1', '2010-1-1', lags=7)
```

```python
lrbt.plot_results()
```

```python
lrbt.optimize_parameters((1,10,1))
```

```python
lrbt.run_strategy('2010-1-1', '2017-12-31',
                  '2018-1-1', '2019-12-31', lags=5)
```

```python
lrbt.plot_results()
```
