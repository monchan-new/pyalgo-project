```python
import LRVectorBacktester as LR
```

```python
lrbt = LR.LRVectorBacktester('USD_JPY', '2002-5-6', '2010-1-1', 10000)
```

```python
lrbt.run_strategy('2002-5-6', '2005-1-1',
                  '2005-1-1', '2007-1-1', lags=3)
```

```python
lrbt.plot_results()
```
