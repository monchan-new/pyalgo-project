```python
import LRVectorBacktester as LR
```

```python
lrbt = LR.LRVectorBacktester('USD_JPY', '2002-5-6', '2010-12-31', 10000)
```

```python
lrbt.run_strategy('2002-5-6', '2005-12-31',
                  '2006-1-1', '2009-12-31', lags=3)
```

```python
lrbt.plot_results()
```

```python
lrbt.optimize_parameters((1,50,1))
```

```python
lrbt.run_strategy('2002-5-6', '2005-12-31',
                  '2006-1-1', '2009-12-31', lags=34)
```

```python
lrbt.plot_results()
```
