```python
from NeuralNetworkVectorBacktester import NeuralNetworkVectorBacktester as NNBT
```

```python
feature_config = {
    "lag": [1,2,3,4,5],
    "momentum": [5,10],
    "volatility": [20],
    "sma": [20,50],
    "ema": [],
    "range": [14]
}
```

```python
nn = NNBT(
    symbol="EUR=",
    start="2010-01-01",
    end="2019-12-31",
    amount=10000,
    tc=0.002,
    lags=5,
    feature_config=feature_config,
    threshold_long=0.55,
    threshold_short=0.45,
    neurons=32,
    epochs=25,
    lr=0.0005
)
```

```python
nn.run_strategy('2010-1-1', '2018-12-31',
                '2019-1-1', '2019-12-31')
```

```python
nn.plot_results()
```

```python
nn.run_strategy('2010-1-1', '2014-12-31',
                '2015-1-1', '2016-12-31')
```

```python
nn.plot_results()
```
